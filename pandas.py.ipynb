# DF Using List

```python
l=[['reddy','brahma','venkata'],[23,35,46]]
l
```

```python
df=pd.DataFrame(data=l)
df
```

```python
l=np.array(l).T
l
```

```python
df=pd.DataFrame(data=l,columns=['Name','Age'])
df
```

# sql
